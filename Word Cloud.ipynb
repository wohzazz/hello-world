```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
# % matplotlib inline
```

# example 

```python
words_to_plot = 'happy to see that'

text = words_to_plot
# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

# example, source data list

```python
words_to_plot = ['muffins', 'cake', 'apple', 'apple', 'apple', 'blueberry'] # list needs to converted to a string. see documentation here: https://www.simplilearn.com/tutorials/python-tutorial/list-to-string-in-python#:~:text=To%20convert%20a%20list%20to%20a%20string%2C%20use,to%20string%20using%20list%20comprehension%20is%20given%20below.
words_to_plot = ' '.join(words_to_plot)
print(words_to_plot)

text = words_to_plot
# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
